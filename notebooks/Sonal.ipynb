```r
plot_afsc = function(datafile, name) {
    meta = read.delim("SraRunTable.txt")
    meta$species = sub(" RAD.*$", "", meta$Organism_s)
    afsc = as.matrix(read.delim(datafile, row.names=1))
    m = match(sub("_il.fastq", "", colnames(afsc)), meta$Run_s)
    afsc = afsc[m,m]
    rownames(afsc) = colnames(afsc) = meta$Organism_s
    afsc.d = as.dist(afsc)
    p = cmdscale(afsc.d)
    plot.df = data.frame(PCo1 = p[,1], PCo2=p[,2], loc=meta$isolation_source_s, org=meta$Organism_s, sp=meta$species)
    p = ggplot(plot.df, aes(x=PCo1, y=PCo2, colour=loc, shape=sp)) +
        ggtitle(name) +
        geom_point(size=2) + 
        theme_classic()
    print(p)
    return(afsc.d)
}
```
